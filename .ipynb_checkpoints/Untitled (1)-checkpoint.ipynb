```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from functools import reduce
from matplotlib.pyplot import figure
```

```python
lot_17 = pd.read_csv('Lottery_FY_17.csv')
lot_18 = pd.read_csv('Lottery_FY_18.csv')
lot_19 = pd.read_csv('Lottery_FY_19.csv')
lot_20 = pd.read_csv('Lottery_FY_20.csv')
lot_21 = pd.read_csv('Lottery_FY_21.csv')



data = {'year': [2017,2018, 2019, 2020, 2021], 
        'total_amount': [lot_17.iloc[-1,-1], lot_18.iloc[-1,-1],lot_19.iloc[-1,-1],lot_20.iloc[-1,-1],lot_21.iloc[-1,-1]]}

sales = pd.DataFrame.from_dict(data)

sales['year'] = sales['year'].astype(str)

sales['total_amount'] = sales['total_amount'].str.replace(",", '').str.replace('$', '').astype(float)



sns.lineplot(data=sales, x="year", y="total_amount",marker="o")
```

```python
fig = px.line(sales, x="year", y="total_amount")
fig.update_traces(textposition="bottom right")
fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 2017,
        dtick = 1
    )
)

fig.show()
```

```python
lot_17['year'] = 2017
lot_18['year'] = 2018
lot_19['year'] = 2019
lot_20['year'] = 2020
lot_21['year'] = 2021
```

```python
lot_17 = lot_17.rename(columns={'FY 17 Net Sales' : 'sales'})
lot_18 = lot_18.rename(columns={'FY 18 Net Sales' : 'sales'})
lot_19 = lot_19.rename(columns={'FY 19 Net Sales' : 'sales'})
lot_20 = lot_20.rename(columns={'FY 20 Net Sales' : 'sales'})
lot_21 = lot_21.rename(columns={'FY 21 Net Sales' : 'sales'})

data_list = [lot_17, lot_18, lot_19, lot_20, lot_21]
result = pd.concat(data_list)
```

```python
fig = px.strip(result, x='sales', y='year')
fig.update_layout(
    yaxis = dict(
        tickmode = 'linear',
        tick0 = 2017,
        dtick = 1
    )
)
fig.show()
```

```python
methuen_lot = result[result['Agent City'] == 'METHUEN'].groupby(['year']).sum().reset_index()
methuen_lot = methuen_lot.drop(methuen_lot[methuen_lot['year'] == 2021].index)
methuen_lot
```
