```python
ml_edu = tabb_mass[['municipal','acs_year','nohs_p', 'lths_p', 'hs_p', 'assoc_p', 'bapl_p', 'mast_p', 'prof_p', 'doc_p']]
ml_edu['municipal'] = ml_edu['municipal'].str.lower()
```

```python

labor['Labor Force'] = labor['Labor Force'].str.replace(',','').astype('float')
```

```python
result['Agent City'] = result['Agent City'].str.lower()
result = result.rename(columns={"sales": "Five Year Lottery Sales"})
```

```python
sales_total = result.groupby(['Agent City']).sum().drop(['year'], axis=1)
```

```python
un_em_rate = labor.groupby(['DOR Code']).mean().drop(['Year'], axis=1)
un_em_rate
```

```python
two = dor.merge(un_em_rate, left_on = 'DOR Code', right_on = 'DOR Code', how = "left")
two['Municipality'] = two['Municipality'].str.lower()
three = two.merge(sales_total, left_on = "Municipality", right_on = "Agent City", how = "left")
four = three.merge(ml_edu, left_on = "Municipality", right_on = 'municipal', how = "left")
```

```python
four.head()
```

```python
four.drop('municipal', axis=1, inplace=True)
four['Population'] = four['Population'].str.replace(',','').str.replace('$','').astype('float')
four['regression target'] = four['Five Year Lottery Sales'] / four['Population']
```

```python
def scatter(x,fig):
    plt.subplot(5,2,fig)
    plt.scatter(four[x],four['regression target'])
    plt.title(x+' vs Lottery purchase per capita')
    plt.ylabel('Lottery purchase per capita')
    plt.xlabel(x)

plt.figure(figsize=(10,20))

scatter('Unemployment Rate', 1) # positive cor seems
scatter('DOR Income Per Capita', 2) # non linear 
scatter('bapl_p', 3) # negative cor
scatter('lths_p', 4)
scatter('Labor Force', 5)
scatter('DOR Code', 6)
# positive cor

plt.tight_layout()
```

```python
df_regression = four[['regression target','Unemployment Rate','DOR Income Per Capita','lths_p','bapl_p', 'Labor Force']]
df_regression = df_regression[df_regression['regression target'].notna()]
```

```python
len(df_regression)
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df_regression, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_vars = ['regression target','Unemployment Rate','DOR Income Per Capita','bapl_p', 'lths_p', 'Labor Force']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
# regression
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
y_train = df_train.pop('regression target')
X_train = df_train
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
    
def checkVIF(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```

```python
X_train_new = build_model(X_train,y_train)
```

```python
checkVIF(X_train_new)
```
